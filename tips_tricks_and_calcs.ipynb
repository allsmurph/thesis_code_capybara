# Hill radii

### Orbital parameters (on exoplanet archive):

| Object   |    Mass               | Radius                          | $a$  (au)      |  $e$                   | $\omega$ (deg)      | $i$ (deg)         |
| :------: | :------:              | :-------:                       | :-----:        | :-----:                | :-----:             | :-----:           |
| PDS 70   | $0.76±0.02$ $M_\odot$ |  $1.26 \pm 0.15 R_{\odot}$      |  --            |       --               | --                  | --                |
| PDS 70 b | $3 ± 1 M_{Jup}$       | $2.72^{+0.20}_{-1.9} R_{Jup}$   | $20 ± 2 $      | $0.19^{+0.15}_{-0.09}$ | $148 ± 31$          | $ 140^{+ 7}_{-6}$ |
| PDS 70 c | $2 ± 1$   $M_{Jup}$   |  $2.04^{+0.61}_{-0.45} R_{Jup}$ | $34^{+6}_{-3}$ |$ 0.11^{+0.12}_{-0.06}$ | $ 136^{+50}_{- 58}$ | $132 ± 7$         |

see https://exoplanetarchive.ipac.caltech.edu/overview/PDS%2070#star_PDS-70_collapsible for more details.


```python
import astropy.constants as const
import numpy as np
import astropy.units as u

def r_hill(a, m, M):
    return a * (m / (3 * M))**(1/3)

hillb = (r_hill(21*u.au, 1*const.M_jup, const.M_sun).to(u.au))
hillc = (r_hill(34*u.au, 6*const.M_jup, const.M_sun).to(u.au))
hilld = (r_hill(11*u.au, 0.4*const.M_jup, const.M_sun).to(u.au))

print(hillb, hillc, hilld)
```

# Kepler III

```python
def period_law(a, M_star):
    return np.sqrt((4 * np.pi**2 * a**3) / (const.G * M_star))

period_law(0.1*u.au, 0.965*const.M_sun).to(u.yr)
```
